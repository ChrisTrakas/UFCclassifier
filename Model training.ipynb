```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import Normalizer,MinMaxScaler
from sklearn.model_selection import learning_curve
from sklearn.model_selection import validation_curve

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
finalTable = pd.read_csv('/home/chris/anaconda3/UFC FINAL DATA.csv')
```

```python


for val in ["AB","BA"]:
    one = val[0]
    two = val[1]
    finalTable['SignStr%'+val] = finalTable.apply(lambda row: row['Sig.str.%avg'+one]*row['Sig.str.def%avg'+two], axis=1)
    finalTable['TD'+val] = finalTable.apply(lambda row: row['TD%avg'+one]*row['TDdef%avg'+two] , axis=1)           
    finalTable['KO'+val] = finalTable.apply(lambda row: row['KOPerc'+one]*row['beenKOPerc'+two] , axis=1)
    finalTable['SUB'+val] = finalTable.apply(lambda row: row['SubPerc'+one]*row['beenSubPerc'+two] , axis=1)                                                    

classes = finalTable[['Class','StanceA','StanceB','FighterA','FighterB']]
finalTable= finalTable.drop(columns=['isWomen','FighterA','FighterB','Class','StanceA','StanceB'])


for col in finalTable.drop(columns=['winnerA']).columns:
    finalTable[col+"2"] = [ el**2 for el in finalTable[col]]

classdummies = pd.get_dummies(classes)
finalTable.join(classdummies)

y = finalTable.winnerA.values
x = finalTable.drop(columns=['winnerA'])

scaler = MinMaxScaler(feature_range=(0, 1))
x = scaler.fit_transform(x)
scaler = Normalizer().fit(x)
x = scaler.transform(x)

X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.33, random_state=42)

```

```python

logisticRegr = LogisticRegression(solver = 'lbfgs' ,max_iter=10000)
logisticRegr.fit(X_train, y_train)
predictions = logisticRegr.predict(X_test)
score = logisticRegr.score(X_test, y_test)
print(score)
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(30,30,20), max_iter=5000)
mlp.fit(X_train,y_train)

predict_test = mlp.predict(X_test)

from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,predict_test))
print(classification_report(y_test,predict_test))
```

```python
#Create a Gaussian Classifier
clf=RandomForestClassifier(n_estimators=500)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

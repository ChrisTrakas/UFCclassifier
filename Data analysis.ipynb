```python
corrmap = sb.diverging_palette(2020,10,as_cmap=True)
sb.heatmap( finalTable[['num_fightsA','perc_winsA' ,'Class','HeightA','KO%A','Subm%A','Sig.str.%avgA','Sig.str.perRoundA' ,'TD%avgA', 'Sub.attperRoundA','PassperRoundA','TDperRoundA']].corr() , vmax=0.3 , center=0 , cmap=corrmap , square=True , linewidths=0.5 , cbar_kws={"shrink": 0.5} )
```

```python
corrmap = sb.diverging_palette(2020,10,as_cmap=True)
sb.heatmap( finalTable[['num_fightsA','perc_winsA' ,'Class','HeightA','beenKO%A','beenSubm%A','Sig.str.def%avgA','TDdef%avgA']].corr() , vmax=0.3 , center=0 , cmap=corrmap , square=True , linewidths=0.5 , cbar_kws={"shrink": 0.5} )
```

```python
graph = sb.barplot(x=finalTable.HeightA , y=finalTable.Class)
graph.set_xticklabels(graph.get_xticklabels() , rotation =90)
plt.show()
```
